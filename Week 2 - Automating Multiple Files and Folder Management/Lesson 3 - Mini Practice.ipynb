# Mini Practice

Build a batch processor that:  
1. Creates a folder called ServerLogs
2. creates at least 3 CSV files inside it
3. Adds at least 1 non-CSV file
4. Loops through the folder
5. Processes only .csv files
6. Prints each row
7. Counts:  
   - total servers
   - down servers
   - high CPU servers over 80

```python
import os
import csv

folder = "ServerLogs1"
total_servers = 0
down_count = 0


if not os.path.exists(folder):
    os.mkdir(folder)

print("Folder is ready:", folder)
print()

sample_files = {
    "WEB01.csv": [
        {"server": "WEB01", "status": "Up", "cpu": 43},
        {"server": "WEB02", "status": "Up", "cpu": 100}
    ],
    "DB01.csv": [
        {"server": "DB01", "status": "Down", "cpu": 37},
        {"server": "DB02", "status": "Up", "cpu": 83}
    ],
    "APP01.csv": [
        {"server": "APP01", "status": "Down", "cpu": 51},
        {"server": "APP02", "status": "Down", "cpu": 57}
    ]
}

for filename, rows in sample_files.items():
    path = os.path.join(folder, filename)

    with open(path, "w", newline = "") as file:
        writer = csv.DictWriter(file, fieldnames=["server", "status", "cpu"])
        writer.writeheader()
        writer.writerows(rows)

print("Sample CSV files created.")
print()

with open(os.path.join(folder, "README.txt"), "w") as file:
    file.write("Ignore this file for testing purposes.")

with open(os.path.join(folder, "notes.md"), "w") as file:
    file.write("This is a markdown file.")

print ("Non-CSV files created.")
print()

for item in os.listdir(folder):
    print(item)
    print()

for file in os.listdir(folder):
    if file.endswith(".csv"):
        print("Processing", file)
        print()

for file in os.listdir(folder):
    if file.endswith(".csv"):
        path = os.path.join(folder, file)

        print("Processing", file)
        print()

        with open(path, "r") as csv_file:
            reader = csv.DictReader(csv_file)

            for row in reader:
                total_servers += 1

                if row["status"] == "Down":
                    down_count += 1
                    
                cpu = int(row["cpu"])
                
                if cpu > 80:
                    print(row["server"], "has high CPU:", cpu)
                 
                print(row)

        print()
print("Total servers:", total_servers)
print("Down servers:", down_count)
```
