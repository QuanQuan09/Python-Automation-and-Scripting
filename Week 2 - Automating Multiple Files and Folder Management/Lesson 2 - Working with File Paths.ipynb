# Mini Practice  
1. Create a new folder named reports  
2. add a text file called summary.txt inside Reports  
3. Write a Python script that:  
- Prints your current working directory
- Lists everything in the current directory
- Builds the path to summary.txt using os.path.join()
- Prints the resulting path

```python
import os

folder_name = "Reports"
file_name = "summary.txt"

if not os.path.exists("Reports"):
    os.mkdir("Reports")

print("Current Working Directory:")
print(os.getcwd())

print("\nContents:")
for item in os.listdir():
    print(item)

path = os.path.join(folder_name, file_name)

with open(path, "w") as file:
    file.write("Weekly Report")

print("\nFile created at:")
print(path)
```
