```python
# Part 1 - Create Folder

import os

folder = "Reports1"

if not os.path.exists(folder):
    os.mkdir(folder)

print("Reports1 folder is ready.")
```

```python
# Part 2 - Create a Sample Report

import os

report_path = os.path.join("Reports1", "Daily_Report.txt")

with open(report_path, "w") as file:
    file.write("Server Report\n")
    file.write("---------------\n")
    file.write("WEB01 - Up\n")
    file.write("WEB02 - Down\n")
    file.write("DB01 - Up\n")

print("Sample report created.")
```

```python
# Part 3 - Create a Backup Folder

import os

backup_folder = "Backups"

if not os.path.exists(backup_folder):
    os.mkdir(backup_folder)

print("Backup folder created.")
    
```

```python
# Part 4 - Copy One File

import os
import shutil

source = os.path.join("Reports1", "Daily_Report.txt")
destination = os.path.join("Backups", "Daily_Report.txt")

shutil.copy(source, destination)

print("Backup completed.")
```

```python
# Part 5 - Verify the Backup

import os

print("Reports1 Folder:")
print(os.listdir("Reports1"))

print()

print("Backups Folder:")
print(os.listdir("Backups"))
```

```python
# Part 6 - Create Multiple Reports

import os
reports = [
    "Monday_Report.txt",
    "Tuesday_Report.txt",
    "Wednesday_Report.txt"
]

for report in reports:

    path = os.path.join("Reports1", report)

    with open(path, "w") as file:
        file.write(f"This is {report}")

print("Reports created.")
```

```python
# Part 7 - Backup Every Report Automatically

import os
import shutil

source_folder = "Reports1"
backup_folder = "Backups"

for file in os.listdir(source_folder):

    source = os.path.join(source_folder, file)

    # Check if the item is a file (not a directory) before copying
    if os.path.isfile(source):
        destination = os.path.join(backup_folder, file)
        shutil.copy(source, destination)
        print("Copied", file)

    else:
        # Skip directories like .ipynb_checkpoints
        print("Skipped directory:", file)
```

```python
# Part 8 - Timestamped Backups

import os
import shutil
from datetime import datetime

timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")

source = os.path.join("Reports1", "Daily_Report.txt")

destination = os.path.join("Backups", f"Daily_Report_{timestamp}.txt")

shutil.copy(source, destination)

print("Timestamped backup created.")
```
