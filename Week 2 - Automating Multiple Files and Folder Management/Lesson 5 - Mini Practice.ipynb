# Mini Practice

This program should:  
1. Create a folder called RenamePractice
2. Create five text files with simple names like:
   - Report1.txt
   - Report2.txt
   - Report3.txt
   - Report4.txt
   - Report5.txt
3. Display the original filenames
4. Rename every file to:
   - ServerReport_001.txt
   - ServerReport_002.txt
   - ServerReport_003.txt
   - ServerReport_004.txt
   - ServerReport_005.txt
5. Display the new filenames
6. Create two additional files with spaces in their names:
   - Server Log.txt
   - Daily Report.txt
7. Rename those files so the spaces become underscores
8. Display the final contents of the folder

```python
import os

folder = "RenamePractice1"

if not os.path.exists(folder):
    os.mkdir(folder)



files = [
    "Report1.txt",
    "Report2.txt",
    "Report3.txt",
    "Report4.txt",
    "Report5.txt",
    "Server Log.txt",
    "Daily Report.txt"
]

for file in files:
    path = os.path.join(folder, file)

    with open(path, "w") as f:
        f.write("Practice file")

print("Original Files:")
print(os.listdir(folder))
print()

count = 1

for file in os.listdir(folder):

    old_path = os.path.join(folder, file)

    if os.path.isfile(old_path) and file.endswith(".txt"):
        new_filename = f"ServerReport_{count:03}.txt"
        new_path = os.path.join(folder, new_filename)
        
        os.rename(old_path, new_path)
        
        print(f"Renamed {file} -> {new_filename}")
        count += 1

print()
print("Renamed files:")
print(os.listdir(folder))
```
