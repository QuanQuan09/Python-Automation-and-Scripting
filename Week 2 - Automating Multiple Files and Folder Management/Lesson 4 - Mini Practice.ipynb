# Mini Practice

My script should:  
1. Create a folder named Reports
2. Create a folder named Backups
3. Generate 3 text files inside Reports
4. Automatically copy every file into Backups
5. Print a message for every copied file
6. Create one timestamped backup of one report
7. Display the contents of both folders using os.listdir()

```python
import os
import shutil
from datetime import datetime

folder = "Reports2"
source_folder = "Reports2"
backup_folder = "Backups1"
report_path = os.path.join("Reports2", "Daily_Report.txt")
reports = [
    "Friday_Report.txt",
    "Saturday_Report.txt",
    "Sunday_Report.txt"
]


if not os.path.exists(folder):
    os.mkdir(folder)

if not os.path.exists(backup_folder):
    os.mkdir(backup_folder)
    
print("Reports2 folder is ready.")
print("Backups1 folder is ready.")
print()


with open(report_path, "w") as file:
    file.write("Daily Server Report\n")
    file.write("-------------------------\n")
    file.write("WEB02 - Down\n")
    file.write("APP03 - Up\n")
    file.write("DB04 - Bouncing\n")

print("Sample report created.")
print()

for report in reports:

    path = os.path.join("Reports2", report)

    with open(path, "w") as file:
        file.write(f"This is {report}")

print("Reports created.")
print()

for file in os.listdir(source_folder):

    source = os.path.join(source_folder, file)

    if os.path.isfile(source):
        destination = os.path.join(backup_folder, file)
        shutil.copy(source, destination)
        print("Copied", file)

    else:
        print("Skipped directory:", file)

print("Reports2 Folder:")
print(os.listdir("Reports2"))
print()
print("Backups1 Folder:")
print(os.listdir("Backups1"))
print()

timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
source = os.path.join("Reports2", "Daily_Report.txt")
destination = os.path.join("Backups1", f"Daily_Report_{timestamp}.txt")
shutil.copy(source, destination)
print("Timestamped backup created.")
```
