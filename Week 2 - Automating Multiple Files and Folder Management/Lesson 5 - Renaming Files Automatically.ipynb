```python
# Part 1 - Create a Practice Folder

import os

folder = "RenamePractice"

if not os.path.exists(folder):
    os.mkdir(folder)

print("Practice folder ready.")
```

```python
# Part 2 - Create Sample Files

import os

folder = "RenamePractice"

files = [
    "Report1.txt",
    "Report2.txt",
    "Report3.txt"
]

for file in files:
    path = os.path.join(folder, file)

    with open(path, "w") as f:
        f.write("Practice file")

print("Files created.")
```

```python
# Part 3 - View the Folder

import os

folder = "RenamePractice"

print(os.listdir(folder))
```

```python
# Part 4 - Rename One File

import os

folder = "RenamePractice"

old_name = os.path.join(folder, "Report1.txt")

new_name = os.path.join(folder, "Monday_Report.txt")

os.rename(old_name, new_name)

print("File renamed.")
```

```python
# Check Folder Again

print(os.listdir(folder))
```

```python
# Part 5 - Rename Every File Automatically

import os

folder = "RenamePractice"

count = 1

for file in os.listdir(folder):

    old_path = os.path.join(folder, file)

    new_filename = f"ServerReport_{count}.txt"

    new_path = os.path.join(folder, new_filename)

    os.rename(old_path, new_path)

    print(f"Renamed {file} -> {new_filename}")

    count += 1
```

```python
# Check Folder

print(os.listdir(folder))
```

```python
# Part 6 - Add Leading Zeros

import os

folder = "RenamePractice"

count = 1

for file in os.listdir(folder):

    old_path = os.path.join(folder, file)

    new_filename = f"ServerReport_{count:03}.txt"

    new_path = os.path.join(folder, new_filename)

    os.rename(old_path, new_path)

    print(new_filename)

    count += 1
```

```python
# Part 7 - Add a Prefix

import os

folder = "RenamePractice"

for file in os.listdir(folder):

    old_path = os.path.join(folder, file)

    new_name = "OLD_" + file

    new_path = os.path.join(folder, new_name)

    os.rename(old_path, new_path)

    print(new_name)
```

```python
# Part 8 - Clean Up Filenames

import os

folder = "RenamePractice"

for file in os.listdir(folder):

    clean_name = file.replace(" ", "_")

    old_path = os.path.join(folder, file)

    new_path = os.path.join(folder, clean_name)

    os.rename(old_path, new_path)

    print(clean_name)
```
