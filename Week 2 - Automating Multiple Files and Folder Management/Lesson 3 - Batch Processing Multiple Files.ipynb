```python
# Create a Fresh Practice Folder

import os

folder = "ServerLogs"

if not os.path.exists(folder):
    os.mkdir(folder)

print("Folder is ready:", folder)
```

```python
# Create Sample CSV Files

import os
import csv

folder = "ServerLogs"

sample_files = {
    "WEB01.csv": [
        {"server": "WEB01", "status": "Up", "cpu": 35},
        {"server": "WEB02", "status": "Down", "cpu": 92}
    ],
    "DB01.csv": [
        {"server": "DB01", "status": "Up", "cpu": 48},
        {"server": "DB02", "status": "Down", "cpu": 88}
    ],
    "APP01.csv": [
        {"server": "APP01", "status": "Up", "cpu": 42},
        {"server": "APP02", "status": "Up", "cpu": 44}
    ]
}

for filename, rows in sample_files.items():
    path = os.path.join(folder, filename)

    with open(path, "w", newline = "") as file:
        writer = csv.DictWriter(file, fieldnames=["server", "status", "cpu"])
        writer.writeheader()
        writer.writerows(rows)

print("Sample CSV files created.")
```

```python
# Add Non-CSV Files to help practice filtering

folder = "ServerLogs"

with open(os.path.join(folder, "README.txt"), "w") as file:
    file.write("This file should not be processed")

with open(os.path.join(folder, "notes.md"), "w") as file:
    file.write("These are notes, not CSV data.")

print("Non-CSV files created.")
```

```python
# Read Every CSV File

import os
import csv

folder = "ServerLogs"

for file in os.listdir(folder):
    if file.endswith(".csv"):
        path = os.path.join(folder, file)

        print("Processing", file)

        with open(path, "r") as csv_file:
            reader = csv.DictReader(csv_file)

            for row in reader:
                print(row)

        print()
```

```python
# Count Total Servers Across All CSV Files

import os
import csv

folder = "ServerLogs"

total_servers = 0

for file in os.listdir(folder):
    if file.endswith(".csv"):
        path = os.path.join(folder, file)

        with open(path, "r") as csv_file:
            reader = csv.DictReader(csv_file)

            for row in reader:
                total_servers += 1

print("Total servers:", total_servers)
```

```python
# Count Down Servers

import os
import csv

folder = "ServerLogs"
down_count = 0

for file in os.listdir(folder):
    if file.endswith(".csv"):
        path = os.path.join(folder, file)

        with open(path, "r") as csv_file:
            reader = csv.DictReader(csv_file)

            for row in reader:
                if row["status"] == "Down":
                    down_count += 1

print("Down servers:", down_count)
```

```python
# Find High CPU Servers

import os
import csv

folder = "ServerLogs"

for file in os.listdir(folder):
    if file.endswith(".csv"):
        path = os.path.join(folder, file)

        with open(path, "r") as csv_file:
            reader = csv.DictReader(csv_file)

            for row in reader:
                cpu = int(row["cpu"])

                if cpu > 80:
                    print(row["server"], "has high CPU:", cpu)
```
