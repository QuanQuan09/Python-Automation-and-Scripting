```python
# Creating a File

with open("daily_report.txt", "w") as file: # Open/create a file
    file.write("Daily Server Report\n") # Text is created inside text file
    file.write("All systems operational.\n")
```

```python
# Reading a file

with open("daily_report.txt", "r") as file:
    contents = file.read()

print(contents)
```

```python
# Appending Information - adding a new line w/o deleting existing content

with open("daily_report.txt", "a") as file:
    file.write("No critical alerts detected.\n")
```

```python
with open("daily_report.txt", "r") as file:
    print(file.read())
```
