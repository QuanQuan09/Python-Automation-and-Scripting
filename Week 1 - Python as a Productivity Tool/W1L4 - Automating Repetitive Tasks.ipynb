```python
# Build a Simple Automation
# suppose we want to count text files

import os

count = 0

for file in os.listdir():
    if file.endswith(".txt"):
        count += 1

print("Total text files:", count)
```

```python
# Automation Example 2
# Build an automatic report

import os

with open("folder_report.txt", "w") as report:

    for file in os.listdir():

        report.write(file + "\n")

print("Report created.")
```

```python
# Automation Example 3
# only text files in report

import os

with open("text_files_report.txt", "w") as report:

    for file in os.listdir():

        if file.endswith(".txt"):
            report.write(file + "\n")

print("Finished!")
```

```python
# Mini Project
# create a script called daily_inventory.py

import os

count = 0

with open("daily_inventory_report.txt", "w") as inventory:

    for file in sorted(os.listdir()):

        if file.endswith(".txt"):
            inventory.write(file + "\n")
            count += 1

print("Completed, total number of files are:", count)

    
```
