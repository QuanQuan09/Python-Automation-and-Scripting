# Project Goal  
Build a File Inventory Generator that:  
- Scans a folder
- Identifies every file
- Categorizes files by type
- Counts each file type
- Generates a professional report

```python
# Step 1 - Import the Module

import os

# Step 2 - Create Counters

count = 14
txt_count = 0
csv_count = 0
pdf_count = 0
ipynb_count = 0
folder_count = 0
other_count = 0


# Step 3 - Create the Report

with open("file_inventory_report.txt", "w") as report:
    report.write("File Inventory Report\n")
    report.write("Adding Files to Inventory\n")

# Step 4 - Loop Through Every File
    
    for file in sorted(os.listdir()):
        report.write(file + "\n")

# Step 5 - Determin the File Type

        if os.path.isdir(file):
            folder_count += 1
            
        if file.endswith(".txt"):
            txt_count += 1

        elif file.endswith(".csv"):
            csv_count += 1

        elif file.endswith(".pdf"):
            pdf_count += 1

        elif file.endswith(".ipynb"):
            ipynb_count += 1

        else:
            other_count += 1

    report.write("\n")
    report.write("Find file\n")
    report.write("Organize file\n")

    report.write(f"TXT Files: {txt_count}\n")
    report.write(f"CSV Files: {csv_count}\n")
    report.write(f"PDF Files: {pdf_count}\n")
    report.write(f"Other Files: {other_count}\n")

print("Inventory report created successfully!")
print(f"Total TXT files: {txt_count}\n")
print(f"Total CSV files: {csv_count}\n")
print(f"Total PDF files: {pdf_count}\n")
print(f"Total Jupyter notebooks: {ipynb_count}\n")
print(f"Total folders: {folder_count}\n")
print(f"Total other files: {other_count}\n")


    
    

```
